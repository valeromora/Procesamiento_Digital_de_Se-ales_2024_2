# ¿Qué sucede si en la represetnación matricial de Fourier se aumenta el tamaño de la matriz?
* Se incrementa la resolución en el dominio de la frecuencia.
* Se pueden capturar más detalles en la transformación.
* Si los nuevos datos no están disponibles, se puede rellenar con ceros (zero-padding), lo que mejora la interpolación en el dominio de la frecuencia.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una matriz de ejemplo 4x4 con valores aleatorios
N = 4
matrix = np.random.rand(N, N)

# Aplicar la FFT a la matriz original
fft_original = np.fft.fft2(matrix)

# Aumentar el tamaño a 8x8 con zero-padding
M = 8
padded_matrix = np.zeros((M, M))
padded_matrix[:N, :N] = matrix  # Insertar la matriz original en la esquina superior izquierda

# Aplicar la FFT a la matriz aumentada
fft_padded = np.fft.fft2(padded_matrix)

# Mostrar los resultados
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("FFT Original (4x4)")
plt.imshow(np.log1p(np.abs(fft_original)), cmap='gray')

plt.subplot(1, 2, 2)
plt.title("FFT con Zero-Padding (8x8)")
plt.imshow(np.log1p(np.abs(fft_padded)), cmap='gray')

plt.show()
```

# ¿Qué sucede si en la represetnación matricial de Fourier se reduce el tamaño de la matriz?
* Se reduce la resolución en el dominio de la frecuencia.
* Se pierde información, lo que puede provocar aliasing.
* Se puede hacer recorte (cropping) de la matriz original o aplicar un filtro para evitar distorsiones.

```python
# Recortar la matriz a 2x2 (se pierde información)
M_small = 2
cropped_matrix = matrix[:M_small, :M_small]

# Aplicar la FFT a la matriz reducida
fft_cropped = np.fft.fft2(cropped_matrix)

# Mostrar los resultados
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("FFT Original (4x4)")
plt.imshow(np.log1p(np.abs(fft_original)), cmap='gray')

plt.subplot(1, 2, 2)
plt.title("FFT Recortada (2x2)")
plt.imshow(np.log1p(np.abs(fft_cropped)), cmap='gray')

plt.show()
```
